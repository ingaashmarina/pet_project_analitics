
# Шаг 3: Понимание признаков
(Одномерный анализ)
- Построение графика признаков
- Распределения

```python
df['Year Introduced'].value_counts()
```

```python
ax = df['Year Introduced'].value_counts() \
.head(20) \
.plot(kind= 'bar', title = 'Top 20 years Coaster Introduced')
ax.set_xlabel('Year Introduced') ,
ax.set_ylabel ('No of Coasters')
```

```python
ax = df['Height(ft)'].plot(kind = 'hist', bins = 20, title = "Coaster Height" )
ax.set_xlabel("Height")
```

```python
ax = df['Height(ft)'].plot(kind = 'kde', title = "Coaster Height" )
ax.set_xlabel("Height")
```


# Шаг 4: Взаимосвязи признаков
- Диаграмма рассеяния
- Корреляция тепловой карты
- Парная диаграмма
- Сравнения по группировке

```python
df.plot(kind = 'scatter',
       x = "Inversions",
        y = "Height(ft)",
title = "Inversions vs Heights")

plt.show()
```

```python
ax = sns.scatterplot(x = "Height(ft)",
                     y = "Inversions",
                    hue = "Year Introduced",
                    data = df)
ax.set_title = ("Coaster Height vs Inversions")

plt.show()
```

```python
sns.pairplot(df,
             vars=['Year Introduced','Latitude',
                   'Height(ft)','Inversions','Gforce'],
            hue='Type Main')
plt.show()
```

```python
df_corr = df[['Year Introduced','Latitude',
    'Height(ft)','Inversions','Gforce']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True)
```
